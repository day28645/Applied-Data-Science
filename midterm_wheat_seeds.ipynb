### Correlation Matrix

```python
# Compute the correlation matrix
numeric_data = df.select_dtypes(include=[np.number])
numeric_data = numeric_data.apply(pd.to_numeric, errors='coerce')  # Convert to NaN if non-numeric
numeric_data = numeric_data.dropna(axis=1, how='all')  # Drop columns where all values are NaN

corr = numeric_data.corr()

# Create the heatmap
f, ax = plt.subplots(figsize=(10, 7))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
_ = sns.heatmap(corr, cmap="YlGn", square=True, ax=ax, annot=True, linewidth=0.1)
plt.title('Pearson Correlation of Features', y=1.05, size=15)
plt.show()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, 7].values
```

## Spilt Train Set and Test Set

```python
from sklearn.model_selection import train_test_split
train_ratio = 147
validation_ratio = 42
test_ratio = 21

# train is now 70% of the entire data set
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 210 - train_ratio)

# test is now 10% of the initial data set
# validation is now 20% of the initial data set
x_val, x_test, y_val, y_test = train_test_split(x_test, y_test, test_size = validation_ratio)
print("train dataset = ", x_train.shape, y_train.shape)
print("test dataset = ", x_test.shape, y_test.shape)
print("validation dataset = ", x_val.shape, y_val.shape)
```
