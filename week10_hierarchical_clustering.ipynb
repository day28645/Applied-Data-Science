# Clustering by Sklearn Library

## Agglomerative Clustering

### Standardization by Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

### Apply PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
X_pca.shape
```

## Parameters : linkage = single

```python
from sklearn.cluster import AgglomerativeClustering

groups_single = AgglomerativeClustering(n_clusters=10, metric='euclidean', linkage='single', compute_distances=True)

groups_single.fit_predict(X_pca)
```

```python
plt.figure(figsize=(10, 10))

for label in dataset.target_names:
  plt.scatter(X_pca[groups_single.labels_ == label, 0],
                  X_pca[groups_single.labels_ == label, 1],
                  label=label)

# plt.scatter(*X_pca.T, c=groups_s.labels_, cmap='viridis')
# plt.colorbar()
plt.title('linkage = single')
plt.legend(loc='upper left')
plt.show()
```

```python
unique, counts = np.unique(groups_single.labels_, return_counts=True)

print(np.asarray((unique, counts)).T)
```

```python
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_, counts]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

```python
plt.figure(figsize=(12, 7))
plt.title('linkage = single (truncate_mode=\'lastp\', p=30)')
plot_dendrogram(groups_single, truncate_mode="lastp", p=30)
plt.xlabel("Index of Point (or number of points in node if parenthesis).")
plt.show()
```
